# 1 whileでできること

while文は処理の繰り返しを行うときに使う。

同じ計算を複数回行うときにwhile文を使うことは基本であるが、書式を理解し工夫することで様々な場面で使用することができる。

例えば、下記にようなことができる。



*   while~else:基本的には繰り返し処理を行うが、特定の条件だけ別の処理を行うことができる。
*   break:繰り返し処理を行うが、特定の条件の時に「break」を記述することで、繰り返し処理から抜けることができる。
*   無限ループ:永遠に処理を繰り返すことができる。例えば、「1.不具合が発生したらその時点でプログラムを停止する」や、「2.無限ループで何らかの事象を待ち、事象が発生したらbreakを使って無限ループを抜けて処理を進める」のような時に使用される。



**while文とfor文の違い**

while文と同様の動作をするものでfor文がある。

どちらも繰り返し処理を行うために使用される点では同じだが、繰り返し条件が異なる。

それぞれの繰り返し条件は以下である。



*   while文:条件を満たしている間
*   for文:指定された回数の間



使い方としては、繰り返し処理の回数がわかっている場合はfor文を使い、繰り返す回数は分からないが、条件を満たす場合に処理をやめたい場合は、while文を使用する。

# 2 具体例

**文法**

```python
while(条件式):
    繰り返す処理
```

Pythonでは、「(条件式)」の後ろに「:」が付くことに注意せよ。

「:」を付け忘れるだけで、プログラムが動作しなくなってしまう。

では、具体的なソースコードでPython while文を組んでみます。

```python
num = 0
while (num < 5):
    print(num)
    num +=1
```

まず、ソースコード1行目で変数numに0を代入している。

そして、ソースコード2行目からwhile文に入る。

while文は条件式(今回は「num < 5」)がTrueである限りwhile以下の処理を続ける。

したがって、処理を行うたびに「num += 1」により変数numに1を足している。

while文の条件式が「num < 5」なので、4までが出力されている。

# 3 while~elseについて

以上のwhile文は、条件式がTrueの間はwhile内の処理が繰り返し実行されるが、条件式がFalseになると、while文は終了する。

しかし、while内で「else」を記述することにより、条件式がFalseの時にelse以下の処理を実行することができる。

それでは、具体例を見ることにする。

```python
num = 0
while(num < 5):
    print(num)
    num +=1
else:
    print("繰り返し処理終了")
```

上記のソースコードでは、変数num=5となる時にwhileの条件式がFalseとなるので、else以下の処理が実行される。

よって、出力結果としては「4」の後に「繰り返し処理終了」という文字が表示されてwhile文全体も終了している。

# 4 while文のbreakについて

Python while文においてbreakを使用すると、break以降の処理を行わずwhile文全体が終了する。

それでは、具体例を見ることにする。

```python
num = 10
while(True):
    print(num)
    num -= 1
    if (num == 0):
        break
```

ソースコード4行目より、変数numが１ずつ減ることがわかる。

そして、ソースコード5行目でif文が登場する。

もし変数numが0になったらbreakを実行することでwhile文を終了する。
その結果、numは1まで減少するところまでで処理が終わり、10~1が出力された。

# 5 while文の無限ループについて

最後に、while文における無限ループについて紹介する。

無限ループとはその名の通り、繰り返し処理が永遠に行われることである。

例えば、以下のソースコードを見てみよう。

```python
num = 0
while (num < 5):
    print(num)
```

上記の場合、変数numに対して数字を足す処理が何もされていないので、変数numはずっと0のままである。

while文の条件式「num < 5」より、0は5よりも小さいので、無限に繰り返し処理を実行し続ける。

よって無限に0が表示された。
無限ループを止めるには、キーボードでCtrl+Cを押すべし。

無限ループはPCへの負荷が大きいので、自宅のノートパソコン等でやる場合は注意せよ。
